```python
import os
import torch
from PIL import Image
import torchvision
import torchvision.transforms as transforms
from torch.autograd import Variable
import torch.nn.functional as F
from net import AGNet
from PIL import ImageFont, ImageDraw
```

```python
net = AGNet()
```

```python
state = torch.load("checkpoint/ckpt.pth", map_location='cpu')
```

```python
net.load_state_dict(state['net'])
```

```python
state[]
```

```python
transform_image = transforms.Compose([
    transforms.CenterCrop(150),
    transforms.ToTensor(),
    transforms.Normalize((0.485,0.456,0.406), (0.229,0.224,0.225))
])
```

```python
path = "/home/damvantai/Documents/projects/predict_age_gender/data/test_2/"
font = ImageFont.truetype("/home/damvantai/Downloads/Aaargh.ttf", 16)

for filename in os.listdir(path):
    img = Image.open(path + filename)
    draw = ImageDraw.Draw(img)
    input_img = transform_image(img)
    input_img.unsqueeze_(0).shape
    age, gender = net(Variable(input_img))
    gender_preds = F.sigmoid(gender)
    print(gender_preds)
    gender_preds = (gender_preds > 0.5).int()
    if gender_preds == 0:
        label = "male"
    if gender_preds == 1:
        label = "female"
    print(label)  
    draw.text((0, 0), label, (255,0,255), font=font)
    img.save(filename)
```
